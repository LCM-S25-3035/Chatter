```python
import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Load your parquet chunk data
PARQUET_PATH = r"C:\Users\choud\Downloads\wiki_hybrid_chunks_300 (2).parquet"

def load_parquet_chunks(parquet_path):
    try:
        df = pd.read_parquet(parquet_path)
        df["chunk_id"] = df["chunk_id"].astype(str)
        return df
    except Exception as e:
        raise RuntimeError(f"Failed to load Parquet file: {e}")

# 2. Load sentence transformer model
def load_embedding_model():
    model = SentenceTransformer("all-MiniLM-L6-v2")
    return model

# 3. Build FAISS index
def build_faiss_index(embeddings):
    embedding_dim = embeddings.shape[1]  # 384 for all-MiniLM-L6-v2
    index = faiss.IndexFlatL2(embedding_dim)
    index.add(embeddings)
    return index

def main():
    print("Loading parquet chunks...")
    df = load_parquet_chunks(PARQUET_PATH)

    print("Loading embedding model...")
    model = load_embedding_model()

    texts = df["chunk_text"].tolist()
    print(f"Encoding {len(texts)} chunks...")
    embeddings = model.encode(texts, convert_to_numpy=True, batch_size=64, show_progress_bar=True)

    print("Building FAISS index...")
    index = build_faiss_index(embeddings)

    print("Ready to answer your queries! Type 'exit' to quit.\n")

    while True:
        query = input("Your question: ").strip()
        if query.lower() == "exit":
            print("Goodbye!")
            break

        query_vec = model.encode([query], convert_to_numpy=True)
        k = 5
        distances, indices = index.search(query_vec, k)

        print("\nTop relevant chunks:\n")
        for idx in indices[0]:
            chunk = df.iloc[idx]
            print(f"Chunk ID: {chunk['chunk_id']}")
            print(f"Text: {chunk['chunk_text']}\n---")

if __name__ == "__main__":
    main()
```
