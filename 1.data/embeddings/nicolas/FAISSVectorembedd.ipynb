```python
import pandas as pd

# Example format: one text chunk per row
df = pd.read_csv(r"C:\Users\gadep\Desktop\dynamic_chunked_output2.csv")  # or however you saved it
texts = df["text"].tolist()  # assuming 'text' column has your chunks
```

```python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer("all-MiniLM-L6-v2")  # Fast and good

embeddings = embedder.encode(texts, batch_size=32, show_progress_bar=True)
```

```python
import faiss
import numpy as np

dimension = 384  # For MiniLM
index = faiss.IndexFlatL2(dimension)

embedding_matrix = np.array(embeddings).astype("float32")
index.add(embedding_matrix)

# Save index to disk
faiss.write_index(index, "faiss_index.index")
```

```python
query_emb = np.array(embedder.encode(["Explain reinforcement learning"])).astype("float32")
distances, indices = index.search(query_emb, k=5)

for idx in indices[0]:
    print(texts[idx])
```
