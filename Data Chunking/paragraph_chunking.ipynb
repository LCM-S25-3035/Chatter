```python
import pandas as pd

# Load the CSV file
file_path = r"C:\Users\choud\Downloads\merged_cleaned_text_PaddleOCR2.csv"
df = pd.read_csv(file_path)

# Print column names to identify the text column
print(df.columns)

# Replace 'text' with the actual column name containing your text
text_column = 'text'

# Drop rows where the text is NaN
df = df.dropna(subset=[text_column])

# Define paragraph chunking function
def chunk_by_paragraph(text):
    # Normalize line breaks
    normalized = str(text).replace('\r\n', '\n').replace('\r', '\n')
    # Split on two or more newlines
    paragraphs = [para.strip() for para in normalized.split('\n\n') if para.strip()]
    return paragraphs

# Apply paragraph chunking
df['paragraph_chunks'] = df[text_column].apply(chunk_by_paragraph)

# Explode so each paragraph appears in a new row
df_exploded = df.explode('paragraph_chunks').reset_index(drop=True)

# Save the result to a new CSV
output_path = r"C:\Users\choud\Downloads\paragraph_chunks_output_v2.csv"
df_exploded.to_csv(output_path, index=False)

print(f"Paragraph-chunked file saved to: {output_path}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer

# Download NLTK data (only once)
try:
    nltk.download('punkt')
except Exception as e:
    print(f"Error downloading NLTK data: {e}")

# Load CSV
csv_path = r"C:\Users\choud\Downloads\paragraph_chunks_output.csv"
df = pd.read_csv(csv_path)

# Assume the text column is named 'text' (change if different)
texts = df['text'].astype(str).tolist()

# Step 1: Compute TF-IDF vectors and cosine similarity matrix
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(texts)

cosine_sim_matrix = cosine_similarity(tfidf_matrix)

print("Cosine Similarity Matrix:")
print(cosine_sim_matrix)

# Step 2: Calculate ROUGE and BLEU between pairs (example: first and second text)

# Initialize ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)

# Example: compare first and second chunk
if len(texts) >= 2:
    reference = texts[0]
    candidate = texts[1]

    # ROUGE scores
    rouge_scores = scorer.score(reference, candidate)
    print("\nROUGE scores between first and second text:")
    print(rouge_scores)

    # BLEU score
    reference_tokens = [nltk.word_tokenize(reference.lower())]
    candidate_tokens = nltk.word_tokenize(candidate.lower())
    bleu_score = sentence_bleu(reference_tokens, candidate_tokens)
    print("\nBLEU score between first and second text:")
    print(bleu_score)
else:
    print("Not enough text chunks to compare.")

```
