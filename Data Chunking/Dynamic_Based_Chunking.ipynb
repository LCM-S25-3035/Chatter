```python
import pandas as pd
import spacy

# Loading spaCy English model
nlp = spacy.load("en_core_web_sm")

# Loading cleaned dataset
df = pd.read_csv(r"C:\Users\gadep\Downloads\merged_cleaned_text_Tesseract.csv")

# Setting the column that contains the cleaned text
text_column = "text"  # Replace with your actual column name

# Define dynamic chunking function
def dynamic_chunk_text(text, max_tokens=512):
    doc = nlp(text)
    chunks = []
    current_chunk = []
    current_length = 0

    for sent in doc.sents:
        sent_tokens = [token.text for token in sent if not token.is_space]
        sent_len = len(sent_tokens)

        # If adding this sentence exceeds max_tokens, finalize current chunk
        if current_length + sent_len > max_tokens:
            if current_chunk:
                chunks.append(" ".join(current_chunk))
            current_chunk = sent_tokens
            current_length = sent_len
        else:
            current_chunk.extend(sent_tokens)
            current_length += sent_len

    # Adding any remaining sentence as a final chunk
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Applying the chunking function to each row in dataset
df["dynamic_chunks"] = df[text_column].apply(lambda x: dynamic_chunk_text(x, max_tokens=512))

# Preview the result
print(df[["dynamic_chunks"]].head())

#Save to a new CSV
df.to_csv(r"C:\Users\gadep\Desktop\dynamic_chunked_output.csv", index=False)

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
import nltk

nltk.download('punkt')

df = pd.read_csv(r"C:\Users\gadep\Desktop\dynamic_chunked_output2.csv")
df.columns = df.columns.str.strip()

vectorizer = TfidfVectorizer()

all_texts = df['text'].tolist() + df['dynamic_chunks'].tolist()
vectorizer.fit(all_texts)

cosine_scores = []
bleu_scores = []
rouge1_scores = []
rouge2_scores = []
rougeL_scores = []

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
smooth_fn = SmoothingFunction().method1

for ref, pred in zip(df['text'], df['dynamic_chunks']):
    ref_vec = vectorizer.transform([ref])
    pred_vec = vectorizer.transform([pred])
    cosine_sim = cosine_similarity(ref_vec, pred_vec)[0][0]
    cosine_scores.append(cosine_sim)

    ref_tokens = nltk.word_tokenize(ref.lower())
    pred_tokens = nltk.word_tokenize(pred.lower())
    bleu = sentence_bleu([ref_tokens], pred_tokens, smoothing_function=smooth_fn)
    bleu_scores.append(bleu)

    scores = scorer.score(ref, pred)
    rouge1_scores.append(scores['rouge1'].fmeasure)
    rouge2_scores.append(scores['rouge2'].fmeasure)
    rougeL_scores.append(scores['rougeL'].fmeasure)

# printing the values

print("Cosine Similarity scores:")
print(cosine_scores)

print("\nBLEU scores:")
print(bleu_scores)

print("\nROUGE-1 scores:")
print(rouge1_scores)

print("\nROUGE-2 scores:")
print(rouge2_scores)

print("\nROUGE-L scores:")
print(rougeL_scores)

# Then add scores to dataframe
df['cosine_similarity'] = cosine_scores
df['bleu'] = bleu_scores
df['rouge1'] = rouge1_scores
df['rouge2'] = rouge2_scores
df['rougeL'] = rougeL_scores

print(df)


```
