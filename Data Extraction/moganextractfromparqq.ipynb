```python
import os
import pandas as pd
from PIL import Image
import pytesseract
import io
import base64

def extract_text_from_parquet_files(parquet_files, output_file, image_column):
    all_text = []

    # Loop through each file in the parquet_files list
    for file_path in parquet_files:
        try:
            # Read the Parquet file
            df = pd.read_parquet(file_path)
            print(f"Processing: {file_path}")
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            continue

        # Check if the image_column exists in the DataFrame
        if image_column not in df.columns:
            print(f"Column '{image_column}' not found in {file_path}. Skipping.")
            continue

        # Iterate over each row in the image_column
        for i, img_data in enumerate(df[image_column]):
            try:
                # Handle image data (either bytes or base64 encoded string)
                if isinstance(img_data, bytes):
                    image = Image.open(io.BytesIO(img_data))
                elif isinstance(img_data, str):  # If the image is base64 encoded
                    image_bytes = base64.b64decode(img_data)
                    image = Image.open(io.BytesIO(image_bytes))
                else:
                    print(f"Unsupported image format at row {i} in {file_path}")
                    continue

                # Extract text using Tesseract OCR
                text = pytesseract.image_to_string(image)
                all_text.append(text)
            except Exception as e:
                print(f"Failed to process image in {file_path}, row {i}: {e}")

    # Save all extracted text to the output file
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("\n\n".join(all_text))

    print(f"✅ Text extraction complete. Output saved to {output_file}")

# Define the list of Parquet files
parquet_files = [
    r"D:\parkfil\0000.parquet",
    r"D:\parkfil\0001.parquet",
    r"D:\parkfil\0002.parquet",
    r"D:\parkfil\0003.parquet",
    r"D:\parkfil\0004.parquet",
    r"D:\parkfil\0005.parquet",
    r"D:\parkfil\0006.parquet",
    r"D:\parkfil\0007.parquet",
    r"D:\parkfil\0008.parquet",
    r"D:\parkfil\0009.parquet",
]

# Specify the output text file name
OUTPUT_TEXT_FILE = r"extracted7_text.txt"  # Saves it in the current working directory



# Replace with the actual column name that holds the image data (check your Parquet file columns)
IMAGE_COLUMN = "image_data"  # You need to replace this with the correct column name

# Call the function to start text extraction
extract_text_from_parquet_files(parquet_files, OUTPUT_TEXT_FILE, IMAGE_COLUMN)
```

```python
def extract_text_from_parquet_files(parquet_files, image_column):
    all_text = []

    for file_path in parquet_files:
        try:
            df = pd.read_parquet(file_path)
            print(f" Loaded {file_path} with {len(df)} rows")
        except Exception as e:
            print(f" Error reading {file_path}: {e}")
            continue

        for i, img_data in enumerate(df[image_column]):
            try:
                if isinstance(img_data, dict) and 'bytes' in img_data:
                    image_bytes = img_data['bytes']
                else:
                    print(f" Unexpected format at row {i}")
                    continue

                image = Image.open(io.BytesIO(image_bytes)).convert("RGB")

                # DEBUG: Save sample image
                if i < 3:
                    image.save(f"preview_{i}.jpg")

                text = pytesseract.image_to_string(image)

                if text.strip():
                    print(f" Text at row {i}:\n{text.strip()[:100]}...")
                    all_text.append(text)
                else:
                    print(f" Empty OCR result at row {i}")

            except Exception as e:
                print(f" Error at row {i}: {e}")

    return all_text
OUTPUT_TEXT_FILE = r"C:\Users\gadep\Desktop\extracted11_text.txt"
os.makedirs(os.path.dirname(OUTPUT_TEXT_FILE), exist_ok=True)

extracted_text = extract_text_from_parquet_files(parquet_files, "image")

with open(OUTPUT_TEXT_FILE, 'w', encoding='utf-8') as f:
    f.write("\n\n".join(extracted_text))

print(f" Done. Saved text to {OUTPUT_TEXT_FILE}")
```
