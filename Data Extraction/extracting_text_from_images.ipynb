```python
from PIL import Image
import pytesseract
import os

# If needed, specify the path to your Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Path to the single image file
image_path = r'C:\Users\sneha\Downloads\parquet_file\images_output\image_5_label_0.jpg'
  # replace with your filename

try:
    with Image.open(image_path) as img:
        print(f"[INFO] Reading: {os.path.basename(image_path)}")
        text = pytesseract.image_to_string(img)
        print("[OCR OUTPUT]")
        print(text)
except Exception as e:
    print(f"[ERROR] Could not process the image: {e}")
```

```python


image_folder = r'C:\Users\sneha\Downloads\parquet_file\images_output'
output_folder = os.path.join(image_folder, 'text_output')
os.makedirs(output_folder, exist_ok=True)  # Create output folder if doesn't exist

chunk_size = 100  # Adjust chunk size or set to None to process all at once

# Get list of all .jpg files
all_files = [f for f in os.listdir(image_folder) if f.lower().endswith('.jpg')]

total_files = len(all_files)
print(f"[INFO] Total .jpg files found: {total_files}")

for start in range(0, total_files, chunk_size if chunk_size else total_files):
    chunk_files = all_files[start:start + (chunk_size if chunk_size else total_files)]
    print(f"[INFO] Processing files {start + 1} to {start + len(chunk_files)} ...")

    for filename in chunk_files:
        image_path = os.path.join(image_folder, filename)
        try:
            with Image.open(image_path) as img:
                text = pytesseract.image_to_string(img)
                
                # Save text to .txt file with the same base name as the image
                base_name = os.path.splitext(filename)[0]
                text_file_path = os.path.join(output_folder, f"{base_name}.txt")
                
                with open(text_file_path, 'w', encoding='utf-8') as f:
                    f.write(text)
                
                print(f"[INFO] Saved OCR text for {filename} to {text_file_path}")
        except Exception as e:
            print(f"[ERROR] Skipping {filename} due to: {e}")

print("[INFO] OCR extraction and saving complete.")
```
