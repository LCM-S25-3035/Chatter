```python
import ast
from PIL import Image
import io

# Try just the first image
first_image_str = combined_df.iloc[0]['image']

try:
    image_dict = ast.literal_eval(first_image_str)
    image_bytes = image_dict['bytes']
    img = Image.open(io.BytesIO(image_bytes))
    img.show()  # Will open image viewer
except Exception as e:
    print(f"Error in first image: {e}")
```

```python
import ast
from PIL import Image
import io
import os

def bytes_to_image(image_str):
    try:
        image_dict = ast.literal_eval(image_str)
        image_bytes = image_dict['bytes']
        image = Image.open(io.BytesIO(image_bytes))
        return image
    except Exception as e:
        print(f"[ERROR] Could not convert image: {e}")
        return None

chunk_size = 500  # You can increase later if needed
output_folder = r'C:\Users\sneha\Downloads\parquet_file\images_output'
os.makedirs(output_folder, exist_ok=True)

saved_count = 0

for start in range(0, len(combined_df), chunk_size):
    chunk = combined_df.iloc[start:start + chunk_size]
    for idx, row in chunk.iterrows():
        img = bytes_to_image(row['image'])
        if img is not None:
            try:
                save_path = os.path.join(output_folder, f'image_{idx}_label_{row["label"]}.jpg')
                img.save(save_path, format='JPEG')
                saved_count += 1
                if saved_count % 100 == 0:
                    print(f"[INFO] Saved {saved_count} images...")
            except Exception as e:
                print(f"[ERROR] Failed to save image {idx}: {e}")
```
