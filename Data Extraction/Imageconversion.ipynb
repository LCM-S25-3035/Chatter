```python
import os
import pandas as pd
from PIL import Image
import io
from tqdm import tqdm

# Directory to save images
image_dir = "extracted_images"
os.makedirs(image_dir, exist_ok=True)

# List of parquet files
input_files = [
    r"C:\Users\japje\Documents\B.G\0000.parquet",
    r"C:\Users\japje\Documents\B.G\0001.parquet",
    r"C:\Users\japje\Documents\B.G\0002.parquet",
    r"C:\Users\japje\Documents\B.G\0003.parquet",
    r"C:\Users\japje\Documents\B.G\0004.parquet",
    r"C:\Users\japje\Documents\B.G\0005.parquet",
    r"C:\Users\japje\Documents\B.G\0006.parquet",
    r"C:\Users\japje\Documents\B.G\0007.parquet",
    r"C:\Users\japje\Documents\B.G\0008.parquet",
    r"C:\Users\japje\Documents\B.G\0009.parquet"
]

# Function to extract and save images
def extract_images_from_parquet(file_path):
    file_index = int(os.path.basename(file_path).split('.')[0])
    try:
        df = pd.read_parquet(file_path)
        for i, row in tqdm(df.iterrows(), total=len(df), desc=f"Extracting from {file_path}"):
            try:
                img_bytes = row["image"]["bytes"]
                img = Image.open(io.BytesIO(img_bytes))
                img_filename = f"{file_index:04d}_img_{i:05d}.png"
                img_path = os.path.join(image_dir, img_filename)
                img.save(img_path)
            except Exception as e:
                print(f"[Error] Image {i} in {file_path}: {e}")
    except Exception as e:
        print(f"[Error] Could not read {file_path}: {e}")

# Process all files
for file in input_files:
    extract_images_from_parquet(file)
```
