```python
import pandas as pd
import os

folder_path = r"C:\Users\sneha\Downloads\parquet_file"  

# List and sort all parquet files
parquet_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.parquet')])

# Print number of rows in each file
for file in parquet_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_parquet(file_path)
    print(f"{file}: {len(df)} rows")
```

```python
combined_df = pd.concat([pd.read_parquet(os.path.join(folder_path, file)) for file in parquet_files])
print("Combined DataFrame Shape:", combined_df.shape)
```

```python
# View the first 5 rows
print(combined_df.head())
```

```python
# Get concise summary of the DataFrame
print(combined_df.info())
```

```python
# View basic statistics of numeric columns
print(combined_df.describe())
```

```python
# Save combined_df to a CSV file in the current folder
combined_df.to_csv(r"C:\Users\sneha\Downloads\parquet_file\combined_df.csv", index=False)
print("File saved as combined_df.csv in your current working directory.")
```
