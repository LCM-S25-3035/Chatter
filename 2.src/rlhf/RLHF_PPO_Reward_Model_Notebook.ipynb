```python
#  STEP 1: Imports
import pandas as pd
import json
from datetime import datetime
import os

#  STEP 2: CSV Log File Path
csv_path = "rlhf_feedback_log.csv"
QUESTION_AI = "What is AI?"
#  STEP 3: Function to Collect Feedback for One Query
def collect_feedback_for_query(query, fallback_answer, direct_answer, variant_answer):
    print("\n📌 Compare Responses:\n")
    print("TinyLLaMA:\n", fallback_answer)
    print("\nDirect:\n", direct_answer)
    print("\nVariant:\n", variant_answer)

    selected_best = input("\n🧠 Which answer was best? (TinyLLaMA / Direct / Variant): ").strip()

    # Prepare log entry
    log_entry = {
        "timestamp": datetime.now().isoformat(),
        "query": query,
        "responses": json.dumps({
            "TinyLLaMA": fallback_answer,
            "Direct": direct_answer,
            "Variant": variant_answer
        }),
        "selected_best": selected_best
    }

    # Save to CSV
    df_log = pd.DataFrame([log_entry])
    df_log.to_csv(csv_path, mode='a', header=not os.path.exists(csv_path), index=False)
    print(f"\n Feedback saved successfully for query: '{query}'")

#  STEP 4: 10 Diverse Examples to Label
examples = [
    {
        "query": QUESTION_AI,
        "TinyLLaMA": "AI is like a machine brain that learns from data.",
        "Direct": "Artificial Intelligence allows machines to mimic human behavior.",
        "Variant": "AI refers to machines simulating intelligent actions."
    },
    {
        "query": "What is quantum computing?",
        "TinyLLaMA": "Quantum computing uses quantum bits to process data differently than classical computers.",
        "Direct": "Quantum computers use superposition and entanglement to solve complex problems.",
        "Variant": "It’s a computing approach that explores quantum mechanics for faster solutions."
    },
    {
        "query": "Explain machine learning.",
        "TinyLLaMA": "Machine learning allows computers to improve performance from experience.",
        "Direct": "ML is a subfield of AI that trains models using data.",
        "Variant": "It's a way to program computers to learn patterns automatically."
    },
    {
        "query": "Define deep learning.",
        "TinyLLaMA": "Deep learning is a type of machine learning using many layers.",
        "Direct": "It’s based on neural networks with multiple layers for feature learning.",
        "Variant": "Deep learning mimics the brain using deep neural networks to understand data."
    },
    {
        "query": "What is reinforcement learning?",
        "TinyLLaMA": "RL is about learning by trial and error using rewards.",
        "Direct": "It’s a type of ML where agents learn from actions and consequences.",
        "Variant": "Reinforcement learning optimizes behavior through rewards and penalties."
    },
    {
        "query": "What is NLP?",
        "TinyLLaMA": "NLP helps computers understand human language.",
        "Direct": "Natural Language Processing is used in chatbots, translators, and sentiment analysis.",
        "Variant": "It enables interaction between machines and humans using text or speech."
    },
    {
        "query": "What is a neural network?",
        "TinyLLaMA": "Neural networks are algorithms inspired by how the brain works.",
        "Direct": "They consist of layers that transform input into useful predictions.",
        "Variant": "Neural networks connect nodes to recognize patterns in data."
    },
    {
        "query": "What is supervised learning?",
        "TinyLLaMA": "Supervised learning trains on labeled data to make predictions.",
        "Direct": "It means using inputs and known outputs to train models.",
        "Variant": "Models learn from examples where the answer is already provided."
    },
    {
        "query": "Explain unsupervised learning.",
        "TinyLLaMA": "Unsupervised learning finds hidden patterns without labeled data.",
        "Direct": "It's about grouping or reducing data without knowing the output.",
        "Variant": "Unsupervised models explore structure in data without guidance."
    },
    {
        "query": "What is a chatbot?",
        "TinyLLaMA": "A chatbot is a computer program that talks to humans.",
        "Direct": "Chatbots simulate conversations and answer user questions.",
        "Variant": "They are AI tools that respond using natural language."
    }
]

#  STEP 5: Loop to Collect Feedback
for ex in examples:
    collect_feedback_for_query(ex["query"], ex["TinyLLaMA"], ex["Direct"], ex["Variant"])

#  STEP 6: View Collected Feedback (Optional)
df = pd.read_csv(csv_path)
print("\n📝 Logged Feedback Preview:")
display(df.tail(10))
```

```python
example_queries = [
    QUESTION_AI, "Define NLP.", "Explain supervised learning.",
    "What is reinforcement learning?", "Define a chatbot.",
    "What is quantum computing?", "Explain deep learning.",
    "What is a neural network?", "What is machine learning?"
]

for ex in example_queries:
    fallback_answer = "TinyLLaMA: Answer for " + ex
    direct_answer = "Direct: Answer for " + ex
    variant_answer = "Variant: Answer for " + ex
    collect_feedback_for_query(ex, fallback_answer, direct_answer, variant_answer)
```

```python
# Step 1: Imports
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import joblib
import ast

REWARD_MODEL_FILE = "reward_model.pkl" # defining a constant
TFIDF_VECTOR_FILE = "tfidf_vectorizer.pkl"
# Step 2: Load feedback
df = pd.read_csv("rlhf_feedback_log.csv")

# Step 3: Prepare Data
texts = []
labels = []

for i, row in df.iterrows():
    try:
        response_dict = ast.literal_eval(row["responses"])
        selected = row["selected_best"].strip().lower()

        if selected not in ["tinyllama", "direct", "variant"]:
            continue

        combined_text = (
            "TinyLLaMA: " + response_dict.get("TinyLLaMA", "") + "\n" +
            "Direct: " + response_dict.get("Direct", "") + "\n" +
            "Variant: " + response_dict.get("Variant", "")
        )

        texts.append(combined_text)
        labels.append(selected)
    except:
        continue

# Step 4: TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

# Step 5: Train model
model = LogisticRegression()
model.fit(X, labels)

# Step 6: Save model and vectorizer
joblib.dump(model, REWARD_MODEL_FILE)
joblib.dump(vectorizer, TFIDF_VECTOR_FILE)

print("✅ Reward model and vectorizer saved successfully!")
```

```python
import joblib

# Load correct model
model = joblib.load(REWARD_MODEL_FILE)
vectorizer = joblib.load(TFIDF_VECTOR_FILE)

# Example input
text_input = """TinyLLaMA: AI is a machine brain
Direct: AI mimics human behavior
Variant: AI simulates intelligent action"""

X = vectorizer.transform([text_input])
predicted_label = model.predict(X)[0]

print("✅ Predicted Best Answer:", predicted_label)
```

```python
#  Step: Simulated PPO Loop using reward model
import joblib
import random

# Load your saved reward model and vectorizer
reward_model = joblib.load(REWARD_MODEL_FILE)
vectorizer = joblib.load(TFIDF_VECTOR_FILE)

# Sample query and candidate responses
query = QUESTION_AI
candidates = [
    "AI is like a machine brain that learns from data.",
    "AI allows machines to mimic human behavior.",
    "AI refers to computers making intelligent decisions."
]

# Evaluate each candidate
X = vectorizer.transform(candidates)
probs = reward_model.predict_proba(X)[:, 1]  # Probability of being the best
best_index = probs.argmax()
best_response = candidates[best_index]

# Print output
print(f"\n Query: {query}")
for i, (text, prob) in enumerate(zip(candidates, probs)):
    print(f"\nResponse {i+1}: {text}\nReward: {prob:.4f}")

print(f"\n Selected Best Response: {best_response}")
```
