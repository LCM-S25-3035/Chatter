```python
import pandas as pd

# Load your labeled dataset
df = pd.read_csv(r"C:\Users\sneha\Downloads\chatter_files\text_quality_classifier.csv")  # Columns: ["chunk", "label"]

# Count good vs. bad
label_counts = df["label"].value_counts()
print(label_counts)

# Output example:
# 1    8500  # Good chunks
# 0    1500  # Bad chunks
# Name: label, dtype: int64
```

```python

# Check columns (adjust names if needed)
print(df.columns)  # Expected: ['text', 'manual_label', 'bert_label']
```

```python
# Inspect 10 random "good" (1) and "bad" (0) chunks
print("=== BERT-LABELED 'GOOD' CHUNKS ===")
print(df[df["label"] == 1].sample(10)["chunk_text"].tolist())

print("\n=== BERT-LABELED 'BAD' CHUNKS ===")
print(df[df["label"] == 0].sample(10)["chunk_text"].tolist())
```

```python
# Filter good chunks (label=1) and keep metadata
good_chunks = df[df["label"] == 1][["group_id", "section", "chunk_id", "chunk_text"]]  # Keep IDs for traceability
good_chunks.to_csv(r"C:\Users\sneha\Downloads\chatter_files\good_chunks.csv", index=False)  # Save for RAG retrieval
```

```python
# Extract question-like chunks
prompts = []
for _, row in good_chunks.iterrows():
    chunk = row["chunk_text"]
    # More flexible conditions:
    if ("?" in chunk) or \
       ("explain" in chunk.lower()) or \
       ("how" in chunk.lower()) or \
       ("what" in chunk.lower()):
        prompts.append({
            "group_id": row["group_id"],
            "chunk_id": row["chunk_id"],
            "prompt": chunk.strip()
        })

# Verify prompts before saving
print(f"Found {len(prompts)} prompts. Sample:")
print(prompts[:3])  # Check first 3 entries

# Save only if prompts exist
if prompts:
    pd.DataFrame(prompts).to_csv(r"C:\Users\sneha\Downloads\chatter_files\prompts.csv", index=False)
else:
    print("Warning: No prompts matched the criteria!")
```
