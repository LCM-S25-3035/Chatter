```python
import pandas as pd
import random

# Load the chunk file
df = pd.read_csv(r"C:\Users\sneha\Downloads\wiki_hybrid_chunks_300.csv")

# Sample 50 random chunks 
sample_df = df.sample(n=50, random_state=42).copy()

# Add an empty label column
sample_df['label'] = None

# Start manual labeling
for idx in sample_df.index:
    print("\n" + "-"*60)
    print(f"Chunk ID: {sample_df.loc[idx, 'chunk_id']}")
    print(f"Text:\n{sample_df.loc[idx, 'chunk_text']}\n")
    
    while True:
        label = input("Enter label (1 = Good / 0 = Bad): ").strip()
        if label in ['0', '1']:
            sample_df.at[idx, 'label'] = int(label)
            break
        else:
            print("Invalid input. Please enter 0 or 1.")

# Save to new file
sample_df.to_csv(r"C:\Users\sneha\Downloads\manually_labeled_chunks.csv", index=False)

print("\n Saved to manually_labeled_chunks.csv")
```
